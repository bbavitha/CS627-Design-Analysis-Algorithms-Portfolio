```python
import matplotlib.pyplot as plt

# Data with actual data from the alogorithm  C++ program's output)
list_sizes = [500, 1500, 2500]  # List sizes used in the algorithm
times = [0.00426, 0.009873, 0.015906]  #runtime times in milliseconds for the respective list sizes

# Plot the data
plt.plot(list_sizes, times, marker='o', linestyle='-', color='b')

# Labeling the plot
plt.xlabel('List Size (n)')
plt.ylabel('Time Elapsed (ms)')
plt.title('Runtime of List Reversal Algorithm')

# Display the plot
plt.grid(True)
plt.show()
```
