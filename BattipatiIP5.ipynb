```python
​​Function SolveSudoku(board):​

   # Base case: If the board is​ full, return it as a solution.

    if CheckIfComplete(board):

        return board

    # ​Find the next empty cell with the​ least number of valid options.

    # This heuristic strongly reduces the search space compared to random guessing.

    emptyRow, emptyCol = GetMostConstrainedCell(board)

    #Loop through all valid values for this cell.

    for candidate in PossibleValues(board, emptyRow, emptyCol):

        if CanPlaceValue(board, emptyRow, emptyCol, candidate):

          # Put the candidate value into the cell.

            board[emptyRow][emptyCol] = candidate

            #Try to recursively solve for the rest of the board.

            # Recursive calls become a little more efficient on account of searching in a restrained manner.

            solution = SolveSudoku(board)

            if solution is not None:

                return solution

            # Backtracked cell gets rest about as empty as possible.

            # Using Backtracking prevents delving down the path of invalid configurations further.

            board[emptyRow][emptyCol] = 0

    # If no solution is identified, we return None.

    return None

Function CheckIfComplete(board):

    # Check that at least ​no empty cells​ remain.

    for each cell in board:

        if cell is empty:

            ​return False​

    ​return True​


Function GetMostConstrainedCell(board):

# The one with the least number of options for a valid solution is recognized.

# The MCV heuristic takes the guesswork out by attempting to fill the most restricted cells first.

    minOptions = infinity

    bestCell = None

    for each cell in board:

        if cell is empty:

            options = PossibleValues(​board, cell​.row, cell.col)

            if size ​of options < minOptions:​

                minOptions = size of options

                bestCell = (cell.row, cell.col)

    return bestCell

Function PossibleValues(board, row, col):

    # Determine the set of numbers that can legally fill the cell at (row, col)

    allValues = {​1, 2, 3, 4, 5, 6, 7, 8, 9​}

    return allValues - (​values in the​ same row, column, or 3x3 box as the cell)

Function CanPlaceValue(​board, row, col, value):​

    # Check if placing the value at​ the specified position follows Sudoku's constraints

    return value ​is not present in the​ same row, column, or 3x3 box
```
